```python
import pandas as pd 
```

```python
from datasets import load_dataset
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
import matplotlib.pyplot as plt
```

```python
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])

df['job_posted_month'] = df['job_posted_date'].dt.month

monthly_counts = df['job_posted_month'].value_counts().sort_index()
monthly_counts.plot(kind='line')
plt.title('Number of Job Postings by Month')
plt.xlabel('month')
```

```python
job_counts = df.job_title_short.value_counts()
job_counts.plot(kind='bar')
plt.title('Count of Job Postings')
plt.ylabel("Counts of Postings")
plt.xlabel('')
```

```python
median_salaries = df.groupby('job_title_short')['salary_year_avg'].median()
median_salaries = median_salaries.sort_values(ascending=True)
median_salaries.plot(kind='barh')
plt.ylabel("")
plt.title('Median Salary by Job Title')
plt.xlabel('Salary in USD ($)')
```
